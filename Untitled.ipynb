# Normalizing data and seperaing train set and test set

```python
X = df3.iloc[1:1100].copy()
y = pd.DataFrame(df2['SalePrice'].iloc[1:1100])
X_test = df3.iloc[1100:].copy()
y_test = pd.DataFrame(df2['SalePrice'].iloc[1100:])

min_X = {}
max_X = {}
min_y = np.min(y_test['SalePrice'])
max_y = np.max(y_test['SalePrice'])
for i in X:
    min_X[i] = np.min(X[i])
    max_X[i] = np.max(X[i])
    X[i] = (X[i] - np.min(X[i]))/((np.max(X[i]) - np.min(X[i])))
    X_test[i] = (X_test[i] - np.min(X_test[i]))/((np.max(X_test[i]) - np.min(X_test[i])))
y['SalePrice'] = (y['SalePrice'] - np.min(y['SalePrice']))/(np.max(y['SalePrice']) - np.min(y['SalePrice']))
y_test['SalePrice'] = (y_test['SalePrice'] - np.min(y_test['SalePrice']))/(np.max(y_test['SalePrice']) - np.min(y_test['SalePrice']))
print(X_test)
X = np.array(X, dtype = np.float32)
X_test = np.array(X_test, dtype = np.float32)
```

# Creating a model to train

```python
class Network(nn.Module):
    def __init__(self):
        super().__init__()
        
        self.linear1 = nn.Linear(74, 500)
        self.linear2 = nn.Linear(500, 250)
        self.linear3 = nn.Linear(250, 125)
        self.linear4 = nn.Linear(125, 1)
        
    def forward(self, x):
        y_pred = self.linear1(x).clamp(min=0)
        y_pred = self.linear2(y_pred).clamp(min=0)
        y_pred = self.linear3(y_pred).clamp(min=0)
        y_pred = self.linear4(y_pred)
        return y_pred
```

```python
# convertig data from padas dataframes and series into tensors
X = th.Tensor(X)
y = th.Tensor(y.values)
X_test = th.Tensor(X_test)
y_test = th.Tensor(y_test.values)
```

```python
in_features = X.size()[1]
out_features = 1
reg_model = Network()
reg_model.cuda()
X = X.to(device)
y = y.to(device)
X_test = X_test.to(device)
y_test = y_test.to(device)
```

# Training the model

```python
mse_loss = th.nn.MSELoss(reduction='sum')
lr = 0.0005
prev_loss = 999999999 
optimizer = th.optim.SGD(reg_model.parameters(), lr=lr)
n_epoch = 500
flag = False
train_losses = []
test_losses = []
for i in range(n_epoch):
    # predict model with current regression parameters
    # forward pass (feed the data to model)
    y_pred = reg_model(X)
    # calculate loss function
    step_loss = mse_loss(y_pred, y)
    
    
    # Backward to find the derivatives of the loss function with respect to regression parameters
    # make any stored gradients to zero
    # backward pass (go back and update the regression parameters to minimize the loss)
    optimizer.zero_grad()
    step_loss.backward()  
    if prev_loss- step_loss <= 0.01 and lr > 0.0001 and flag == True:
        for g in optimizer.param_groups:
            g['lr'] = lr -0.00005
            lr = lr -0.00005
            print(lr)
            flag = False
    elif prev_loss - step_loss<= 0.01:
        flag = True
    # update with current step regression parameters 
    optimizer.step()
    prev_loss = step_loss
    train_losses.append(prev_loss.cpu())
    with th.no_grad():
        y_pred_test = reg_model(X_test)
        # calculate loss function
        step_loss_test = mse_loss(y_pred_test, y_test)
        test_losses.append(step_loss_test.cpu())
    print ('epoch [{}], Loss: {:.2f}'.format(i, step_loss.item()))
    print ('epoch [{}], test Loss: {:.2f}'.format(i, step_loss_test.item()))
```

```python
tr_losses= [ loss.detach().numpy() for loss in train_losses]
te_losses= [ loss.detach().numpy() for loss in test_losses]
plt.plot(tr_losses, label='Training loss')
plt.plot(te_losses, label='test loss')
plt.legend(frameon=False)
# Plot to check whether overfitting exists
```
